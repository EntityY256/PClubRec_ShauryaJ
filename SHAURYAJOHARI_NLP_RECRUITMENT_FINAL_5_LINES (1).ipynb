```python
def get_match(query_embedding, top_k=5):
    #We will convert paragraphs and queries to simple vectors for easier comparision
    # paragraphs_embeddings = model.encode(paragraphs, convert_to_tensor=True)

    # Cosine similarities between the query and each paragraph are computed
    cosine_scores = util.pytorch_cos_sim(query_embedding, X)[0]

    # Get the top_k (5) most relevant paragraphs
    top_results = torch.topk(cosine_scores, k=top_k)

    # Extract the paragraphs and scores
    relevant_paragraphs = [(idx, score.item()) for idx, score in zip(top_results.indices, top_results.values)]
    ids=[]
    for i in range(0,top_k):
      S=f"{relevant_paragraphs[i]}"
      S=remove_punctuations(S)
      # print(S)
      ids.append(S[1])
    return ids
```

```python
def get_relevant_paragraphs(query_embedding, Tensor1, Tensor2, index,top_k=5):
    #We will convert paragraphs and queries to simple vectors for easier comparision
    # query_embedding = model.encode(query, convert_to_tensor=True)
    # paragraphs_embeddings = model.encode(paragraphs, convert_to_tensor=True)

    # Cosine similarities between the query and each paragraph are computed
    cosine_scores = util.pytorch_cos_sim(query_embedding, Tensor1)[0] + util.pytorch_cos_sim(query_embedding, Tensor2)[0]

    # Get the top_k (5) most relevant paragraphs
    top_results = torch.topk(cosine_scores, k=top_k)

    # Extract the paragraphs and scores
    relevant_paragraphs = [(idx, score.item()) for idx, score in zip(top_results.indices, top_results.values)]
    ids=[]
    for i in range(0,top_k):
      S=f"{relevant_paragraphs[i]}"
      S=remove_punctuations(S)
      # print(S)
      ids.append(S[1])
    numbers=[]
    lines=[]
    for i in (ids):
      numbers.append(index[int(i)])
    for i in (numbers):
      lines.append(dfp.loc[int(i)].iat[1])

    return lines
```

```python
def get5lines(query="Nintendo Entertainment System discontinued "):

  q = model.encode(query, convert_to_tensor=True)
  BestKeywords= get_match(q, top_k=2)
  W1 = []
  W2 = []
  index=[]
  for i in BestKeywords:
    print(wordlist[int(i)])
    for j in range (dfr.loc[int(i)].iat[1],dfr.loc[int(i)].iat[2]):
      W1.append(Y[j])
      W2.append(Z[j])
      index.append(j)
  M=torch.stack(W1)
  N=torch.stack(W2)
# print(Z.shape)
  LINES=get_relevant_paragraphs(q, M,N,index)
  return (LINES)
```
